#Join Articles + annotations in tsv




```
 def Extract_tag_annotations(df): 
  id_list = []
  tag_list = []
  tag_matrix = []

  for index, row in df.iterrows():
    #print(tag_list)
    if index ==0:
      id_new = row[0]
      id_list.append(id_new)
      tag = row[1]
      tag_list.append(tag)
    else:
      #print(index)
      id_old = id_new
      id_new = row[0]
      tag = row[1]

      if id_new != id_old:
        #print("nuevo: ", a, "antiguo: ", b)
        id_list.append(id_new)
        tag_matrix.append(tag_list)
        tag_list = []
        tag_list.append(tag)

      else:
        tag_list.append(tag)
  tag_matrix.append(tag_list)

  #print(id_list) 
  #print(len(id_list))

  #print(tag_matrix)
  #print(len(tag_matrix))
  return(id_list, tag_matrix)


def Create_transposed_list(list):
  #print("lenght and list are: ", len(list), "\t", list)
  np_list = np.asarray([list]).T
  #print("lenght and transposed token IOB list are: ", len(np_list), "\t", np_list)
  return(np_list)

def Create_tag_dataframe(id_list, np_id_list, np_tag_matrix):  
  df = pd.DataFrame(columns=['article_id','tags'])
  for a in range(len(id_list)):
    df.loc[a] = [np.random.randint(0,2) for n in range(2)]
  for b in range(len(id_list)):
    df.loc[:, ['article_id']] = np_id_list
    df.loc[:, ['tags']] = np_tag_matrix
  #print(tabulate(df, headers= 'keys', tablefmt='psql'))

  return(df)
```

```
def Join_annotation_articles(path, df_annotations, df_articles):
  id_list, tag_matrix = Extract_tag_annotations(df_annotations)
  np_id_list = Create_transposed_list(id_list)
  np_tag_matrix = Create_transposed_list(tag_matrix)
  df = Create_tag_dataframe(id_list, np_id_list, np_tag_matrix)
  df_joined = pd.merge(df_articles, df, on = 'article_id')
  print(df_joined)
  df_joined.to_csv(path, index=False)             #guardar dataframe como tsv sin el indice de la izquierda
  print("---Saved Joined dataframe in: ", path) 
  return(df_joined)
```
